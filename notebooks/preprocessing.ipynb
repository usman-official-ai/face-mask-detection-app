```python
"""
Face Mask Detection - Data Split (Train/Val/Test)
Author: Muhammad Usman
"""

import os
import shutil
import random

# Split ratios
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

source_path = '../dataset/train'
dest_val = '../dataset/val'
dest_test = '../dataset/test'

def split_data(source, dest_val, dest_test):
    for class_name in ['with_mask', 'without_mask']:
        class_path = os.path.join(source, class_name)
        if not os.path.exists(class_path):
            print(f"Class folder not found: {class_path}")
            continue
            
        images = os.listdir(class_path)
        random.shuffle(images)
        
        total = len(images)
        val_end = int(total * val_ratio)
        test_end = int(total * (val_ratio + test_ratio))
        
        val_images = images[:val_end]
        test_images = images[val_end:test_end]
        train_images = images[test_end:]
        
        # Create destination folders
        for folder in [dest_val, dest_test]:
            os.makedirs(os.path.join(folder, class_name), exist_ok=True)
        
        # Move images to val and test
        for img in val_images:
            shutil.move(os.path.join(class_path, img), os.path.join(dest_val, class_name, img))
        for img in test_images:
            shutil.move(os.path.join(class_path, img), os.path.join(dest_test, class_name, img))
        
        # Remaining images stay in train
        remaining = os.listdir(class_path)
        
        print(f"{class_name}: Train={len(remaining)}, Val={len(val_images)}, Test={len(test_images)}")

split_data(source_path, dest_val, dest_test)
print(" Data split complete!")
```
